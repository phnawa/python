```python
df.plot(kind='bar',stacked=True)
```

```python
df['survivied rate']=df.survived/(df.survived+df.dead)
df['dead rate']=df.dead/(df.survived+df.dead)
df
```

```python
df[['survivied rate','dead rate']].plot(kind='bar',stacked=True)
```

```python
tatanic['title']=tatanic.Name.apply(lambda name:name.split(',')[-1].split('.')[0].strip())
tatanic.head()
```

```python
tatanic['family_size']=tatanic.SibSp+tatanic.Parch+1
def func(family_size):
    if family_size<=1:
        return 'single'
    elif family_size>=2 and  family_size<=4:
        return 'small family_size'
    else:
        return 'large family_size'
tatanic['family_type']=tatanic.family_size.apply(func)
tatanic
```

```python
tatanic.family_type.value_counts()
```
